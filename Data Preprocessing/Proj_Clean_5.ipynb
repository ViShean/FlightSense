# Scale Weather Data, save the minb and max values to be used for predictions

```python
import json
weather_cols = [ 
    'STD_temp', 'STD_dwpt', 'STD_rhum', 'STD_prcp', 'STD_snow',
    'STD_wdir', 'STD_wspd', 'STD_wpgt', 'STD_pres', 'STD_tsun', 'STD_coco',
    'ATD_temp', 'ATD_dwpt', 'ATD_rhum', 'ATD_prcp', 'ATD_snow', 'ATD_wdir',
    'ATD_wspd', 'ATD_wpgt', 'ATD_pres', 'ATD_tsun', 'ATD_coco', 'STA_temp',
    'STA_dwpt', 'STA_rhum', 'STA_prcp', 'STA_snow', 'STA_wdir', 'STA_wspd',
    'STA_wpgt', 'STA_pres', 'STA_tsun', 'STA_coco', 'ATA_temp', 'ATA_dwpt',
    'ATA_rhum', 'ATA_prcp', 'ATA_snow', 'ATA_wdir', 'ATA_wspd', 'ATA_wpgt',
    'ATA_pres', 'ATA_tsun', 'ATA_coco' ]
# Step 1: Calculate and store min and max for each column
scale_params = {}
for column in weather_cols:
    scale_params[column] = {
        'min': df[column].min(),
        'max': df[column].max()
    }

# Step 2: Apply Min-Max scaling
for column in weather_cols:
    min_val = scale_params[column]['min']
    max_val = scale_params[column]['max']
    df[column + '_scaled'] = (df[column] - min_val) / (max_val - min_val)

# Show scaled DataFrame
print(scale_params)

with open('weather_scale_params.json', 'w') as file:
    json.dump(scale_params, file, indent=4)
```

# Normalise Datetimes to use sin and cosine, to follow cyclic version of dates

```python
# Follow the sun patterns of the day, start at midnight
def cosine_time_of_day(datetime_series):
    # Calculate fractional hours
    fractional_hours = datetime_series.dt.hour + datetime_series.dt.minute / 60 + datetime_series.dt.second / 3600
    # Cosine transformation, peaking at midnight
    return np.cos(2 * np.pi * fractional_hours / 24)

# Function to calculate cosine for time of year
# Follow hot cold season, where start of the year is cold, middle is hot and end is cold
def cosine_time_of_year(datetime_series):
    day_of_year = datetime_series.dt.dayofyear
    return np.cos(2 * np.pi * day_of_year / 365.25)
```

```python
PLANE_TIMINGS = ['STD_UTC', 'ATD_UTC', 'ATA_UTC', 'STA_UTC']
```

```python
for col in PLANE_TIMINGS:
    df[col + '_time_of_day_cosine'] = cosine_time_of_day(df[col]) # Relative to general heat in a day
    df[col + '_time_of_year_cosine'] = cosine_time_of_year(df[col]) # Relative to general heat in a nothern hemisphere location.
    df[col + '_day_of_year'] = df[col].dt.dayofyear.astype('Int64') # Day number in year (1-365)
    df[col + '_week_of_year'] = df[col].dt.isocalendar().week.astype('Int64')  # Week number (1-52)
    df[col + '_weekday'] = df[col].dt.weekday.astype('Int64')  # 0 = Monday, 6 = Sunday
    df[col + '_hour_of_day'] = df[col].dt.hour.astype('Int64')
```

CONVERT STA STD AND ATD TO DATETIME 24 HOURS FORMAT 

```python


# Columns to convert
timeCols = ['STD', 'STA', 'ATD']

tqdm.pandas(desc='Processing rows')

# Function to convert time
def convert_mixed_time(time):
    if pd.isnull(time) or str(time).strip() == '':
        return None
    try:
        # Attempt 24-hour format
        return pd.to_datetime(time, format='%H:%M', errors='raise').time()
    except:
        try:
            # Attempt 24-hour format with seconds
            return pd.to_datetime(time, format='%H:%M:%S', errors='raise').time()
        except:
            try:
                # Attempt 12-hour format with AM/PM
                return pd.to_datetime(time, format='%I:%M %p', errors='raise').time()
            except:
                return None

# Apply the function to each time column
for col in timeCols:
     df[col] = [convert_mixed_time(t) for t in tqdm(df[col], desc=f"Converting {col}")]   
    
df
```

```python
df.to_csv('SCRAPED_MERGED_DATA_PROCESSED.csv', index=False)
```

Create ATA from Status 

```python
df['ATA'] = (
    df['STATUS']
   .str.extract(r'Landed\s(\d{1,2}:\d{2}(?:\s?(?:AM|PM))?)')[0]  # Extract time
   .apply(convert_mixed_time)  # Convert to time 
   
)
df['ATA'] = df['ATA'].apply(lambda x: x.strftime('%H:%M:%S') if x else None) 
df
```

```python
df.to_csv('SCRAPED_MERGED_DATA_PROCESSED.csv', index=False) 
```

Convert the Age column to Age (in years) so it can be in integer type  
Remove the word "years" and extract only the numeric portion.  
Convert "Brand new" to 0.  
Convert '-' and NaN to None (or NaN, depending on your preference).  
Convert the resulting column to a numeric type.

```python
df['Age'].unique() 
```

```python
# Clean 'Age' column
def clean_age(value):
    if pd.isna(value) or value == '-':
        return None  # Convert NaN or '-' to None
    if value == 'Brand new':
        return 0  # Convert 'Brand new' to 0
    # Remove 'years' and 'year', then convert to integer
    return int(value.split()[0]) if value.split()[0].isdigit() else None

# Apply the cleaning function
df['Age'] = df['Age'].apply(clean_age)

# Convert to numeric dtype for LightGBM compatibility
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df
```

rename Age to Age(Years)

```python
df.rename(columns={"Age": "Age(years)"}, inplace=True)
print(df)

df.to_csv('SCRAPED_MERGED_DATA_PROCESSED.csv', index=False) 
```

Convert FLIGHT TIME TO FLIGHT TIME (mins)

```python
# Ensure the column is string
df['FLIGHT TIME'] = df['FLIGHT TIME'].astype(str)

# Apply the lambda only to rows matching the HH:MM pattern
valid_mask = df['FLIGHT TIME'].str.match(r'^\d{1,2}:\d{2}$', na=False)

valid_mask
```

```python
df.loc[valid_mask, 'FLIGHT TIME'] = df[valid_mask]['FLIGHT TIME'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
df
```

```python
# Convert to float, invalid conversions become NaN
df['FLIGHT TIME'] = pd.to_numeric(df['FLIGHT TIME'], errors='coerce')

# Confirm results
print(df['FLIGHT TIME'].dtype)
print(df['FLIGHT TIME'].isna().sum())
```

```python
import re 

def remove_hidden_chars(text):
        if pd.isna(text):
              return text
        # Remove any character outside the printable ASCII range
        return re.sub(r'[^\x20-\x7E]', '', str(text))

# Apply the cleaning to the entire column
df['Serial Number'] = df['Serial Number'].apply(remove_hidden_chars)

# Verify if hidden characters still exist
for index, val in df['Serial Number'].items():
    valStr = str(val) 
    for c in str(valStr):
        if ord(c) > 127:
            print(f"row {index} : hidden character {c} : {valStr}")


```
